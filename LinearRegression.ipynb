```python
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df["year"], df["per capita income (US$)"],color='red',marker='.')
```

```python
model =linear_model.LinearRegression()
model.fit(df[['year']],df[['per capita income (US$)']])
```

```python
prediction = model.predict(pd.DataFrame([[2018]], columns=['year']))
print(prediction)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(df[['year']],df["per capita income (US$)"])
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df["year"], df["per capita income (US$)"],color='red',marker='.')
plt.plot(df["year"],model.predict(df[["year"]]),color='green')
```
